```python
modelos = [SVC(kernel = 'rbf', C= 100, gamma= 1, decision_function_shape='ovo'),
           SVC(kernel = 'poly', C= 1, gamma= 'scale', decision_function_shape='ovo')
    
]
```

```python
#Metodo para probar diferentes kernel con diferentes muestras (for for)
# print('PRUEBA SATIFACOTRIA DE BUCLES FOR FOR\n')
for samples in range(6250, nSamples+1, 6250):
    print(f'\033[5;35m===================================== {int(samples*(1-nTest))} ========================================\033[0;m')
    df = data
    df = df[:samples]
    X = df.drop(['act'], axis = 'columns')
    y = df.act
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=nTest, random_state=42)
    for model in modelos:
        print(f'\033[5;34mNúmero de muestras: {int(samples * (1-nTest))} // kernel: {model.kernel} // C: {model.C} // gamma: {model.gamma}\033[0;m')
        modelo = model
        list_acc, time, best_model = Modulo_SVM.modl(modelo, X_train, y_train, 10)
        mean_acc = np.mean(list_acc)
        print(f'\033[5;31mAccuracy medio de validación: {mean_acc}\033[0;m')
        print(f'\033[5;32mAccuracy test: {best_model.score(X_test,y_test)} \n\033[0;m')
        
```
