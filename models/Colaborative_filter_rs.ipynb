```python
dataFrame = pd.read_csv(path_absolute+path_file, sep=',')
data = dataFrame.copy()
```

```python
# Especifica el tamaño de la muestra que deseas obtener
tamano_muestra = 8000  # Cambia esto al tamaño de muestra que desees
# Obtiene una muestra aleatoria uniforme del DataFrame
data = data.sample(n=tamano_muestra, random_state=42)  # random_state para reproducibilidad
data.head()
```

```python
import pandas as pd
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import KNNBasic
from surprise import accuracy

# Suponiendo que 'data' es tu DataFrame con las columnas 'userId', 'movieId', 'rating', 'tag', y 'timestamp'
reader = Reader(rating_scale=(1, 5))

# Cargar datos en Surprise usando DatasetAutoFolds
# dataset = Dataset.load_from_df(data[['userId', 'movieId', 'rating', 'tag']], reader)

# Dividir el conjunto de datos en conjuntos de entrenamiento y prueba
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Configurar y entrenar el modelo de KNNBasic (vecinos más cercanos)
model = KNNBasic(sim_options={'user_based': True})  # Filtrado colaborativo basado en usuarios
model.fit(trainset)

# Realizar predicciones en el conjunto de prueba
predictions = model.test(testset)

# Calcular la precisión del modelo
rmse = accuracy.rmse(predictions)
print(f'RMSE: {rmse}')
```
