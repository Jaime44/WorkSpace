```python
df = pd.read_csv(path_absolute+nameDataFrame, sep=',')
print(df.shape)
print(df.info())
df.head()
dataFrame = df.copy()
```

```python
dataFrame.rename(columns={'timestamp_valoraciones': 'timestamp_scr', 'timestamp_etiquetas': 'timestamp_tags'}, inplace=True)
```

```python
# Supongamos que tienes un DataFrame llamado 'data' con una columna 'géneros'
# Ejemplo de cómo se ve la columna 'géneros': 'Acción|Aventura|Ciencia ficción'

# Divide la columna 'géneros' en múltiples columnas usando '|' como delimitador y convierte en binario (0 o 1)
generos_separados = dataFrame['genres'].str.get_dummies(sep='|')

# Concatena los nuevos géneros_separados con el DataFrame original
data = pd.concat([dataFrame, generos_separados], axis=1)

# Elimina la columna original 'géneros' si ya no la necesitas
data = data.drop(columns=['genres'])

# Ahora, las columnas de géneros son variables binarias (0 o 1) que indican la presencia o ausencia de cada género
# print(data.head())

dataFrame = data.copy()
```

```python
columna_a_mover = dataFrame.pop('rating')

# Inserta la columna 'rating' en la última posición del DataFrame
dataFrame.insert(len(dataFrame.columns), 'rating', columna_a_mover)
```

```python
# Supongamos que tienes un DataFrame llamado 'dataFrame' y quieres eliminar las columnas 'tmdbId' y 'imdbId'.
columnas_a_eliminar = ['imdbId', 'tmdbId']

# Elimina las columnas especificadas del DataFrame 'data'
dataFrame = dataFrame.drop(columnas_a_eliminar, axis=1)

# Ahora, tmdbId' y 'imdbId' han sido eliminadas del DataFrame 'dataFrame'
```
